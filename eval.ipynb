# Train models

```python
# Train CTM
model_ctm = CTM(num_topics=n_topics)
output_ctm = model_ctm.train_model(dataset)
```

```python
# Train ETM
model_etm = ETM(num_topics=n_topics, device="gpu")
output_etm = model_etm.train_model(dataset)
```

```python
# Train HDP
model_hdp = HDP()
output_hdp = model_hdp.train_model(dataset)
```

```python
# Train LDA
model_lda = LDA(num_topics=n_topics, iterations=500, passes=5)
output_lda = model_lda.train_model(dataset)
```

```python
# Train LSI
model_lsi = LSI(num_topics=n_topics)
output_lsi = model_lsi.train_model(dataset)
```

```python
# Train NMF
model_nmf = NMF(num_topics=n_topics)
output_nmf = model_nmf.train_model(dataset)
```

```python
# Train NeuralLDA
model_nlda = NeuralLDA(num_topics=n_topics)
output_nlda = model_nlda.train_model(dataset)
```

```python
# Train ProdLDA
model_plda = ProdLDA(num_topics=n_topics)
output_plda = model_plda.train_model(dataset)
```

# Evaluation

```python
# Initialize metrics
topic_diversity = TopicDiversity(topk=10)
topic_coherence = Coherence(texts=dataset.get_corpus(), topk=10, measure='c_npmi')
```

## a) CTM

```python
# Retrieve metrics score
topic_diversity_score = topic_diversity.score(output_ctm)
print("Topic diversity: " + str(topic_diversity_score))

topic_coherence_score = topic_coherence.score(output_ctm)
print("Coherence: " + str(topic_coherence_score))
```
